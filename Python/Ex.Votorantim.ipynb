```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
texto = " ".join(data['Descrição'][~data['Descrição'].isna()])
wordcloud = WordCloud(max_font_size = 80, max_words = 100, width = 1000, height = 1000).generate(texto)
fig = plt.figure(figsize = (100, 20), dpi = 100, facecolor = 'w', edgecolor = 'k')
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
